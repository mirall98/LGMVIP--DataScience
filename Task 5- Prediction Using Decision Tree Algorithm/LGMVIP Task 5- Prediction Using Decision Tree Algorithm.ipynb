### Separating Variables

```python
x = df[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]
y = df[['Species']]
```

```python
x
```

```python
y
```

### Splitting The Data

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
x_train.shape
```

```python
x_train
```

### Building Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier

# Assuming x_train, y_train, x_test, and y_test are defined

model = DecisionTreeClassifier(criterion='gini', min_samples_split=2, random_state=0)
model.fit(x_train, y_train)

y_train_pred = model.predict(x_train)
y_test_pred = model.predict(x_test)
```

```python
model = DecisionTreeClassifier(criterion='gini', min_samples_split=2, random_state=0)
model.fit(x_train, y_train)
y_train_pred = model.predict(x_train)
y_test_pred = model.predict(x_test)
```

```python
tree.plot_tree(model)
```

### Visualizing The Model

```python
from sklearn import tree
import matplotlib.pyplot as plt
```

```python
independent_var = list(df.columns[0:4])
independent_var
```

```python
species = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
fig, axes = plt.subplots(nrows=1 , ncols=1 ,figsize=(5,4),dpi=300)
tree.plot_tree(model,
               feature_names = independent_var,
               class_names = species,
               filled = True,
               node_ids = True,
               fontsize = 3,
              )
```

### Checking For The Accuracy

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
print('Accuracy Score: ' ,accuracy_score(y_test,y_test_pred)*100,'%')
print('Confusion Matrix: \n',confusion_matrix(y_test,y_test_pred))
```

### Predicting With New Data

```python
data = [[5.1,2.3,4.5,7.3],[2.3,6.1,3.2,1.2]]
print(model.predict(data))
```
